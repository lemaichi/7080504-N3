## Yêu cầu 3: Thực hiện thay đổi kích thước của ảnh gốc theo tham số tỷ lệ % nhập vào từ bàn phím. Nếu số nhập vào < 100 (thu nhỏ) sử dụng thuật toán cv2.INTER_AREA, ngược lại nếu số nhập vào >=100 (phóng to) sử dụng thuật toán cv2.INTER_CUBIC. Hiển thị kết quả như hình minh họa dưới và lưu ảnh đã thay đổi vào thư mục Pic/Saves/MaSV_img_scale.jpg.

**Trường hợp nhập vào tỷ lệ <100**

```python
scale = int(input('Nhập tỷ lệ % thay đổi kích thước ảnh:'))
                  

#Tính toán chiều rộng và chiều cao ảnh theo tỷ lệ thiết lập:
w_scale = int(w*scale/100)
h_scale = int(h*scale/100)

dsize = (w_scale,h_scale)

#Thực hiện resize ảnh theo scale:
img_scale_min = cv2.resize(img_meo,dsize, interpolation=cv2.INTER_AREA)

print('Kích thước ảnh gốc:',h,'x',w)
print('Kích thước ảnh gốc sau thay đổi:',scale,'x',w_scale)
#Hiển thị ảnh:
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(img_meo)
plt.title('Ảnh Gốc')

plt.subplot(1, 2, 2)
plt.imshow(img_scale_min)
plt.title('Ănh Đã Scale')

plt.show()
```

```python
cv2.imwrite('C:/Users/Admin/Documents/7080518_Thigiacmaytinh/BaiTest_so1/Pic/Saves/2121050422_img_scale.jpg'
            ,img_scale_min)
```

**Trường hợp nhập vào tỷ lệ >=100**

```python
scale = int(input('Nhập tỷ lệ % thay đổi kích thước ảnh:'))
                  

#Tính toán chiều rộng và chiều cao ảnh theo tỷ lệ thiết lập:
w_scale = int(w*scale/100)
h_scale = int(h*scale/100)

dzoom_out = (w_scale,h_scale)

#Thực hiện resize ảnh theo scale:
img_scale_max = cv2.resize(img_meo,dzoom_out, interpolation=cv2.INTER_CUBIC)

print('Kích thước ảnh gốc:',h,'x',w)
print('Kích thước ảnh gốc sau thay đổi:',scale,'x',w_scale)

#Hiển thị ảnh:
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(img_meo)
plt.title('Ảnh Gốc')

plt.subplot(1, 2, 2)
plt.imshow(img_scale_max)
plt.title('Ănh Đã Scale')

plt.show()
```
